Seaborn
1. seaborn is data visulization library, which is build on the top of the matplotlib to add some functionlity and make it easy to use.
1. best part is it support pandas dataframe.

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
iris.head(3)
```

```python
iris.info()
```

```python
iris.describe(include='object')
```

```python
iris['species'].unique()
```

```python
#how data distributed
#much cleanner then matplotlib, added labels etc
sns.histplot(iris['petal_length'], bins=20)
plt.show()
```

kdf plot - keranl density plot
> give us the distributuion of data

```python
#y-axis is probability, which show petal lenght of 1.7 is 0.20
sns.kdeplot(iris['petal_length'])
plt.show()
```

Scatterplot
> when we increse the petal_length then petal_width also increse

```python
sns.scatterplot(x=iris['petal_length'], y=iris['petal_width'])
plt.show()
```

```python
sns.scatterplot(x='petal_length', y='petal_width', data=iris, hue='species')
plt.show()
```

```python
#let see how hard it to write the above diagram using the matplotlib and aperciate the seaborn.
setosa = iris.loc[iris['species'] == 'setosa']
versicolor = iris.loc[iris['species'] == 'versicolor']
virginica = iris.loc[iris['species'] == 'virginica']

plt.scatter(x=setosa['petal_length'], y=setosa['petal_width'], c='blue')
plt.scatter(x=versicolor['petal_length'], y=versicolor['petal_width'], c='orange')
plt.scatter(x=virginica['petal_length'], y=virginica['petal_width'], c='green')
plt.show()
```

```python
sns.jointplot(x='petal_length', y='petal_width', data=iris, hue='species')
plt.show()
```

```python
sns.pairplot(data=iris, hue='species')
plt.show()
```

### Categorical Data

1. Count plot

```python
sns.countplot(x='species', data=iris)
plt.show()
```

2. Box plot
> if you have one categorical value and one numerical value then use box plot or violin plot

> outliers = 1.5xIQR(middle 50% data)

```python
sns.boxplot(x='species', y='sepal_length', data=iris)
plt.show()
```

3. Violin plot

```python
sns.violinplot(x='species', y='sepal_length', data=iris)
plt.show()
```

4. corelation

```python
iris.corr()
```

```python
sns.heatmap(iris.corr(), cmap='Blues', annot=True)
plt.show()
```
