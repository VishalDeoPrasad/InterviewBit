```python
import numpy as np
x_val = np.arange(-3, 3.1, 0.1)
y_val = x_val**2
plt.plot(x_val, y_val, color="r", marker="*")

plt.title("Test Plot")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.xticks(np.arange(-3, 3.1, 0.2), rotation=90)
plt.yticks(np.arange(0, 9.5, 0.5))

plt.show()
```

Limit your Axes(plot)

```python
import numpy as np
x_val = np.arange(-3, 3.1, 0.1)
y_val = x_val**2
plt.plot(x_val, y_val, color="r", marker="*")

plt.title("Test Plot")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.xlim(0,3)

plt.show()
```

Multiple Plot on One Axes

```python
import numpy as np
x_val = np.arange(-3, 3.1, 0.1)
y_val1 = x_val**2
y_val2 = x_val**3

plt.plot(x_val, y_val1, label="Square")
plt.plot(x_val, y_val2, label="Cube")

plt.text(0,0+2,"(0,0)")
plt.text(1,1+2,"(0,0)")

plt.xticks(np.arange(-3, 3.2, 0.2), rotation=90)
plt.yticks(np.arange(-30, 30, 2))

plt.legend(["*Square","*Cube"], loc="lower right", ncol=2, frameon=False)
plt.grid()
plt.show()
```
