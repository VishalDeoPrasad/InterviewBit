Multiple Plot on One Axes

```python
import numpy as np
x_val = np.arange(-3, 3.1, 0.1)
y_val1 = x_val**2
y_val2 = x_val**3

plt.plot(x_val, y_val1, label="Square")
plt.plot(x_val, y_val2, label="Cube")

plt.text(0,0+2,"(0,0)")
plt.text(1,1+2,"(0,1)")

plt.xticks(np.arange(-3, 3.2, 0.2), rotation=90)
plt.yticks(np.arange(-30, 30, 2))

plt.legend(["*Square","*Cube"], loc="lower right", ncol=2, frameon=True)
plt.grid()
plt.show()
```

Sigmoid function/logistic 


```python
import math
sigmoid = lambda x:1/(1+math.exp(-x))
x = np.arange(-100, 100)
y = [sigmoid(i) for i in x]
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.xlim(-10,10)
plt.show()
```

Subplots

```python
x = np.arange(1, 10, 0.1)
sin_fun = np.sin(x)
tan_fun = np.tan(x)
log_fun = np.log(x)

fig, ax = plt.subplots(3,1) #row and columns
ax[0].plot(x, sin_fun)
ax[1].plot(x, tan_fun)
ax[2].plot(x, log_fun)
plt.show()
```

```python
x = np.arange(1, 10, 0.1)
sin_fun = np.sin(x)
tan_fun = np.tan(x)
log_fun = np.log(x)

plt.figure()

plt.subplot(2,3,1).plot(x, sin_fun)
plt.subplot(2,3,3).plot(x, tan_fun)
plt.subplot(2,3,5).plot(x, log_fun)

plt.show()
```

```python
fig = plt.figure()
fig.add_subplot(231).plot(x, sin_fun)
fig.add_subplot(233).plot(x, tan_fun)
fig.add_subplot(235).plot(x, log_fun)
plt.show()
```

### 3D Plot

```python
a = np.array([0,1,2,3])
b = np.array([0,1,2])
a,b = np.meshgrid(a,b)
print(a)
print(b)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(a,b,np.array([[7]]))
plt.show()
```

```python
a = np.arange(-1,1,0.005)
b = a
a,b = np.meshgrid(a,b)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(a,b, a**2+b**2)
plt.show()
```
